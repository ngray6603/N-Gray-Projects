```python
import nltk
from gensim.models import Word2Vec
from nltk.corpus import gutenberg, webtext

# 1. Download everything we need
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('gutenberg')
nltk.download('webtext')

# 2. LOAD & CLEAN DATA
# We lowercase everything so 'King' and 'king' are the same word
shakespeare_data = [[w.lower() for w in sent] for sent in gutenberg.sents('shakespeare-hamlet.txt')]
modern_data = [[w.lower() for w in sent] for sent in webtext.sents('firefox.txt')] # Firefox help forum text

# 3. TRAIN SHAKESPEARE MODEL (Skip-gram)
# sg=1 means "Skip-gram"
sh_model = Word2Vec(sentences=shakespeare_data, vector_size=100, window=5, min_count=1, sg=1)

# 4. TRAIN MODERN MODEL (Skip-gram)
mod_model = Word2Vec(sentences=modern_data, vector_size=100, window=5, min_count=1, sg=1)

# 5. THE COMPARISON TEST
word = 'fire' # A word that exists in both worlds

print("--- RESULTS ---")
try:
    print(f"In SHAKESPEARE, '{word}' is similar to:", sh_model.wv.most_similar(word, topn=3))
    print(f"In MODERN TEXT, '{word}' is similar to:", mod_model.wv.most_similar(word, topn=3))
except KeyError:
    print(f"The word '{word}' wasn't found in one of the datasets!")
```
